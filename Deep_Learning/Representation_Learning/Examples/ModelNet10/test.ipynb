```python
import gzip
import numpy as np

with gzip.open('../Datasets/ModelNet10/modelnet10.npy.gz', 'rb') as f:
    data = np.load(f)
    print(type(data), data.shape)
```

```python
import matplotlib.pyplot as plt

voxels = data[78][0]

# rotate the voxels
voxels = np.rot90(voxels, axes=(0, 2))
voxels = np.rot90(voxels, axes=(1, 2))


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.voxels(voxels, facecolors='red')
plt.show()
```
