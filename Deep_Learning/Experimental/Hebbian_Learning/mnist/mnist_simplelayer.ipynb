```python
class MixSamples():
    def __init__(self, dataset):
        self.dataset = dataset
        self.n = len(dataset)
    def __call__(self, x, _=None):
        i = torch.randint(0, self.n-1, (x.shape[0],))
        return mix_images(x, self.dataset[i][0])

def mix_images(x1, x2, steps=10):
    device = x1.device
    mask = torch.bernoulli(torch.ones((28,28))*0.5).unsqueeze(0).unsqueeze(0).to(device)
    # blur  with a filter of the form [1/4, 1/2, 1/4] in both horizontal and veritical directions
    filter_h = torch.tensor([[1/4, 1/2, 1/4]]).to(device)
    filter_v = torch.tensor([[1/4], [1/2], [1/4]]).to(device)
    for _ in range(5):
        mask = F.conv2d(mask, filter_h.unsqueeze(0).unsqueeze(0), padding='same')
        mask = F.conv2d(mask, filter_v.unsqueeze(0).unsqueeze(0), padding='same')
    
    # threshold at 0.5
    mask_1 = mask > 0.5
    mask_2 = mask <= 0.5

    out = x1*mask_1 + x2*mask_2
    return out.squeeze(0)
```

```python
def dtanh(x):
    return 1.0 - torch.tanh(x).square()
```

```python
class LayerV1(nn.Module):
    def __init__(self, in_features, out_features):
        super(LayerV1, self).__init__()
        self.in_features = in_features
        self.out_features = out_features

        self.weight = nn.Parameter(torch.empty(out_features, in_features)) 
        self.bias = nn.Parameter(torch.empty(in_features))
        self.reset_parameters()
        self.weight.data *= 0.1

        self.error = None
        self.belief = None

    def update_parameters(self, lr):
        self.weight.data += lr * (torch.tanh(self.belief.T) @ self.error) / self.error.shape[0]
        self.bias.data += lr * self.error.mean(dim=0)
        
    def reset_parameters(self):
        nn.init.kaiming_uniform_(self.weight, a=math.sqrt(5))
        if self.bias is not None:
            fan_in, _ = nn.init._calculate_fan_in_and_fan_out(self.weight)
            bound = 1 / math.sqrt(fan_in) if fan_in > 0 else 0
            nn.init.uniform_(self.bias, -bound, bound)
            
    def init_state(self, batch_size, rand=False):
        if rand:
            self.belief = torch.randn(batch_size, self.out_features).to(self.weight.device) * 0.1
            self.error = torch.randn(batch_size, self.in_features).to(self.weight.device) * 0.1
        else:
            self.belief = torch.zeros(batch_size, self.out_features).to(self.weight.device)
            self.error = torch.zeros(batch_size, self.in_features).to(self.weight.device)
    
    def forward(self, x, y=None, dt=0.01):
        self.error += dt * (x - F.linear(torch.tanh(self.belief), self.weight.T, self.bias))
        self.belief += dt * (F.linear(self.error, self.weight, None))
        if y is not None:
            self.belief = F.normalize(y, dim=1) * torch.linalg.vector_norm(self.belief, dim=1, keepdim=True)
```

```python
# With precision weighted errs
class LayerV2(nn.Module):
    def __init__(self, in_features, out_features):
        super(LayerV2, self).__init__()
        self.in_features = in_features
        self.out_features = out_features

        self.weight = nn.Parameter(torch.empty(out_features, in_features)) 
        self.bias = nn.Parameter(torch.empty(in_features))
        self.var = nn.Parameter(torch.empty(in_features, in_features))
        self.reset_parameters()

        self.error = None
        self.belief = None

    def update_parameters(self, lr, reg=None):
        self.weight.data += lr * (torch.tanh(self.belief.T) @ self.error) / self.error.shape[0]
        self.bias.data += lr * self.error.mean(dim=0)
        self.var.data += lr * (((self.e.T @ self.error) / self.error.shape[0]) - torch.eye(self.in_features).to(self.weight.device))
        if reg is not None:
            self.weight.data -= lr * reg * self.weight.data
            self.var.data -= lr * reg * self.var.data
        
    def reset_parameters(self):
        nn.init.kaiming_uniform_(self.weight, a=math.sqrt(5))
        self.weight.data *= 0.1
        nn.init.kaiming_uniform_(self.var, a=math.sqrt(5))
        self.var.data *= 0.1
        self.var.data += torch.eye(self.in_features)
        if self.bias is not None:
            fan_in, _ = nn.init._calculate_fan_in_and_fan_out(self.weight)
            bound = 1 / math.sqrt(fan_in) if fan_in > 0 else 0
            nn.init.uniform_(self.bias, -bound, bound)
            
    def init_state(self, batch_size, rand=False):
        if rand:
            self.e = torch.randn(batch_size, self.in_features).to(self.weight.device) * 0.1
            self.error = torch.randn(batch_size, self.in_features).to(self.weight.device) * 0.1
            self.belief = torch.randn(batch_size, self.out_features).to(self.weight.device) * 0.1
        else:
            self.e = torch.zeros(batch_size, self.in_features).to(self.weight.device)
            self.error = torch.zeros(batch_size, self.in_features).to(self.weight.device)
            self.belief = torch.zeros(batch_size, self.out_features).to(self.weight.device)
    
    def forward(self, x, y=None, td_err=None, dt=0.01):
        self.e += dt * (F.linear(self.error, self.var, None) - self.e)
        self.error += dt * (x - F.linear(torch.tanh(self.belief), self.weight.T, self.bias) - self.e)
        self.belief += dt * (dtanh(self.belief) * F.linear(self.error, self.weight, None))
        # self.belief += dt * F.linear(self.error, self.weight, None)
        if td_err is not None:
            self.belief -= 0.1 * dt * td_err
        if y is not None:
            self.belief = F.normalize(y, dim=1) * torch.linalg.vector_norm(self.belief, dim=1, keepdim=True)
  
```

```python
# With lateral inhibition
class LayerV3(nn.Module):
    def __init__(self, in_features, out_features):
        super(LayerV3, self).__init__()
        self.in_features = in_features
        self.out_features = out_features

        self.weight = nn.Parameter(torch.empty(out_features, in_features)) 
        self.bias = nn.Parameter(torch.empty(in_features))
        self.var = nn.Parameter(torch.empty(in_features, in_features))
        self.lat = nn.Parameter(torch.empty(out_features, out_features))
        self.reset_parameters()

        self.error = None
        self.belief = None
        
    def reset_parameters(self):
        nn.init.kaiming_uniform_(self.weight, a=math.sqrt(5))
        self.weight.data *= 0.1
        nn.init.kaiming_uniform_(self.var, a=math.sqrt(5))
        self.var.data *= 0.1
        self.var.data += torch.eye(self.in_features)
        nn.init.kaiming_uniform_(self.lat, a=math.sqrt(5))
        self.lat.data *= 0.1
        self.lat.data -= self.lat.data.diag().diag()
        if self.bias is not None:
            fan_in, _ = nn.init._calculate_fan_in_and_fan_out(self.weight)
            bound = 1 / math.sqrt(fan_in) if fan_in > 0 else 0
            nn.init.uniform_(self.bias, -bound, bound)

    def update_parameters(self, lr):
        self.weight.data += lr * (torch.tanh(self.belief.T) @ self.error) / self.error.shape[0]
        self.bias.data += lr * self.error.mean(dim=0)
        self.var.data += lr * (((self.e.T @ self.error) / self.error.shape[0]) - torch.eye(self.in_features).to(self.weight.device))
        dlat = self.belief.T @ self.belief.abs()
        dlat -= dlat.diag().diag()
        self.lat.data += lr * (dlat - self.lat.data)
            
    def init_state(self, batch_size, rand=False):
        if rand:
            self.e = torch.randn(batch_size, self.in_features).to(self.weight.device) * 0.1
            self.error = torch.randn(batch_size, self.in_features).to(self.weight.device) * 0.1
            self.belief = torch.randn(batch_size, self.out_features).to(self.weight.device) * 0.1
        else:
            self.e = torch.zeros(batch_size, self.in_features).to(self.weight.device)
            self.error = torch.zeros(batch_size, self.in_features).to(self.weight.device)
            self.belief = torch.zeros(batch_size, self.out_features).to(self.weight.device)
    
    def forward(self, x, y=None, dt=0.01):
        self.e += dt * (F.linear(self.error, self.var, None) - self.e)
        self.error += dt * (x - F.linear(torch.tanh(self.belief), self.weight.T, self.bias) - self.e)
        self.belief += dt * (F.linear(self.error, self.weight, None) - F.linear(torch.tanh(self.belief), self.lat, None))
        if y is not None:
            self.belief = F.normalize(y, dim=1) * torch.linalg.vector_norm(self.belief, dim=1, keepdim=True)
  
```

```python
class Model(nn.Module):
    def __init__(self, sizes):
        super(Model, self).__init__()
        self.sizes = sizes
        self.layers = nn.ModuleList([LayerV2(sizes[i], sizes[i+1]) for i in range(len(sizes)-1)])

    def update_parameters(self, lr, reg=None):
        for l in self.layers:
            l.update_parameters(lr, reg)

    def init_state(self, batch_size, rand=False):
        for l in self.layers:
            l.init_state(batch_size, rand)

    def update_state(self, x, y=None, dt=0.01):
        for i, l in enumerate(self.layers):
            if i < len(self.layers)-1:
                l(x, td_err=self.layers[i+1].error, dt=dt)
                x = l.belief
            else:
                l(x, y, dt=dt)

    def forward(self, x, y=None, rand=False, max_t=20, dt=0.01):
        self.init_state(x.size(0), rand)

        for _ in range(int(max_t/dt)):
            self.update(x, y, dt)

    @property
    def output(self):
        return self.layers[-1].belief

    @property
    def error_norm(self):
        error = 0
        for l in self.layers:
            error += l.error.norm(dim=1).mean()
        return error / len(self.layers)
```

```python
torch.manual_seed(42)
# Parameter Initialisation
sizes = [784, 256, 10]
max_t = 10
dt = 0.02
model = Model(sizes).to(device)

# Track epochs indexed: [layer][epoch]
epochs_E_mags = [[] for _ in range(len(model.layers))]
epochs_R_mags = [[] for _ in range(len(model.layers))]
epochs_Us_mags = [[] for _ in range(len(model.layers))]
epochs_Us_stds = [[] for _ in range(len(model.layers))]


# Hyperparameters
lr = 0.001
reg_coeff = 10.0
corr_coeff = 1.0
EPOCHS = 10
BATCH_SIZE = 512

val_dataloader = torch.utils.data.DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

track = False
if track:
    E_track = [[[torch.zeros(sizes[layer_i]) for layer_i in range(len(model.layers))] for step_i in range(steps)] for epoch_i in range(EPOCHS)]
    R_track = [[[torch.zeros(sizes[layer_i+1]) for layer_i in range(len(model.layers))] for step_i in range(steps)] for epoch_i in range(EPOCHS)]
    Us_track = [[torch.zeros((sizes[layer_i+1] * sizes[layer_i])) for layer_i in range(len(model.layers))] for epoch_i in range(EPOCHS)]
track_batch = 0

track_weights = True 
if track_weights:
    weights_track = [torch.empty((EPOCHS, sizes[layer_i+1], sizes[layer_i])) for layer_i in range(len(model.layers))]
    for layer_i in range(len(model.layers)):
        weights_track[layer_i][0] = model.layers[layer_i].weight.data.clone()

val_acc = []
val_celoss = []

# Learning Loop
for ep in range(EPOCHS):
    # Track batches indexed: [layer][batch]
    batches_E_mags = [[] for _ in range(len(model.layers))]
    batches_R_mags = [[] for _ in range(len(model.layers))]
    batches_Us_mags = [[] for _ in range(len(model.layers))]
    batches_Us_stds = [[] for _ in range(len(model.layers))]

    epoch_train_celoss = 0

    # for idx in range(len(positive_dataset)):
    pos_dataloader = torch.utils.data.DataLoader(positive_dataset, batch_size=BATCH_SIZE, shuffle=True)
    negative_dataset.apply_transform()
    neg_dataloader = iter(torch.utils.data.DataLoader(negative_dataset, batch_size=BATCH_SIZE, shuffle=True))
    loop = tqdm(enumerate(pos_dataloader), total=len(pos_dataloader), leave=False)
    if ep > 0:
        temp_e = torch.zeros(len(model.layers))
        for i in range(len(model.layers)):
            temp_e[i] = epochs_E_mags[i][-1]
        loop.set_description(f"Epoch {ep+1}/{EPOCHS} - E: {temp_e.mean().item():.3f} - Val CE: {val_celoss[-1]:.3f} - Val Acc: {val_acc[-1]:.3f}")

    for batch_i, (x, y) in loop:
        # Positive Pass
        x = x.flatten(start_dim=1).to(device)
        y = F.one_hot(y, num_classes=sizes[-1]).float().to(device)
        batch_size = x.shape[0]
        model.init_state(batch_size, rand=True)

        with torch.no_grad():
            for _ in range(int(max_t/dt)):
                model.update_state(x, y)
            model.update_parameters(lr, reg_coeff)
        
        # Track batches
        for i, l in enumerate(model.layers):
            batches_E_mags[i].append(l.error.norm(dim=1).mean().item())
            batches_R_mags[i].append(l.belief.norm(dim=1).mean().item())
            # batches_Us_mags[i].append(Us[i].weight.data.abs().mean().item())
            batches_Us_mags[i].append(l.weight.data.mean().item())
            batches_Us_stds[i].append(l.weight.data.std().item())
    
        # Negative Pass
        x, y = next(neg_dataloader)
        x = x.flatten(start_dim=1).to(device)
        y = F.one_hot(y, num_classes=sizes[-1]).float().to(device)
        batch_size = x.shape[0]
        model.init_state(batch_size, rand=True)

        with torch.no_grad():
            for _ in range(int(max_t/dt)):
                model.update_state(x, y)
            model.update_parameters(-1.0*lr)

    # Track epochs
    for i in range(len(model.layers)):
        epochs_E_mags[i].append(torch.tensor(batches_E_mags[i]).mean().item())
        epochs_R_mags[i].append(torch.tensor(batches_R_mags[i]).mean().item())
        epochs_Us_mags[i].append(torch.tensor(batches_Us_mags[i]).mean().item())
        epochs_Us_stds[i].append(torch.tensor(batches_Us_stds[i]).mean().item())

    if track_weights:
        for i, l in enumerate(model.layers):
            weights_track[i][ep] = l.weight.data.clone()
    
    # Eval 
    epoch_val_correct = 0
    epoch_val_celoss = 0
    for x, y in val_dataloader:
        bs = x.shape[0]
        x = x.flatten(start_dim=1).to(device)
        model.init_state(bs, rand=True)

        for step_i in range(int(max_t/dt)):
            with torch.no_grad():
                model.update_state(x, dt=dt)
            
        pred = model.output.argmax(dim=1)
        epoch_val_correct += (pred == y).sum().item()
        epoch_val_celoss += F.cross_entropy(F.softmax(model.output, dim=1), y, reduction='sum').item()

    val_acc.append(epoch_val_correct/len(positive_dataset))
    val_celoss.append(epoch_val_celoss/len(positive_dataset))


# # Plot weights
# plt.figure(figsize=(15, 5))
# for i in range(len(Us)):
#     plt.subplot(1, len(Us), i+1)
#     plt.imshow(Us[i].weight.data.cpu().numpy())
#     plt.colorbar()
# plt.show()

temp_e = torch.zeros(len(model.layers))
for i in range(len(model.layers)):
    temp_e[i] = epochs_E_mags[i][-1]
print(f'Final E: {temp_e.mean().item():.6f}')

# Plot Stats
fig, axs = plt.subplots(2,2, figsize=(8,8))
for i in range(len(model.layers)):
    axs[0,0].plot(epochs_E_mags[i], label=f"Layer {i}")
    axs[0,0].set_title('E_mags')
    axs[0,0].legend()
    axs[0,1].plot(epochs_R_mags[i], label=f"Layer {i}")
    axs[0,1].set_title('R_mags')
    axs[0,1].legend()
    axs[1,0].plot(epochs_Us_mags[i], label=f"Layer {i}")
    axs[1,0].set_title('Weight Mean')
    axs[1,0].legend()
    axs[1,1].plot(epochs_Us_stds[i], label=f"Layer {i}")
    axs[1,1].set_title('Weight Mean Std')
    axs[1,1].legend()
plt.show()

# plot val acc and loss
fig, axs = plt.subplots(1,2, figsize=(8,4))
axs[0].plot(val_acc)
axs[0].set_title('Val Acc')
axs[1].plot(val_celoss)
axs[1].set_title('Val CE Loss')
plt.show()

```

```python
def step2(x, E, R, Us, Vs, y=None, corr_coeff=1.0):
    inp = x
    for i in range(len(Us)):
        E[i] = inp - Us[i](R[i])
        R[i] = R[i] + Vs[i](E[i])
        inp = R[i]
    if y is not None:
        y_norm = y / torch.norm(y)
        y_scaled = y_norm * torch.norm(R[-1])
        R[-1] = (corr_coeff*y_scaled) + ((1-corr_coeff)*R[-1])
    return E, R
```

```python
## Dual Weights U and V

# Parameter Initialisation
sizes = [784, 256, 10]
steps = 4
Us = [nn.Linear(sizes[i+1], sizes[i], bias=True) for i in range(len(sizes)-1)]
Vs = [nn.Linear(sizes[i], sizes[i+1], bias=True) for i in range(len(sizes)-1)]
for i in range(len(Us)):
    Us[i].weight.data *= 0.1
    Us[i] = Us[i].to(device)
    Vs[i].weight.data *= 0.1
    Vs[i] = Vs[i].to(device)

# Track epochs indexed: [layer][epoch]
epochs_E_mags = [[] for _ in range(len(Us))]
epochs_R_mags = [[] for _ in range(len(Us))]
epochs_Us_mags = [[] for _ in range(len(Us))]
epochs_Us_stds = [[] for _ in range(len(Us))]

# Hyperparameters
lr = 0.000034
reg_coeff = 0.2
corr_coeff = 1.0
EPOCHS = 50
BATCH_SIZE = 256

dataloader = torch.utils.data.DataLoader(positive_dataset, batch_size=BATCH_SIZE, shuffle=True)

track = False
if track:
    E_track = [[[torch.zeros(sizes[layer_i]) for layer_i in range(len(Us))] for step_i in range(steps)] for epoch_i in range(EPOCHS)]
    R_track = [[[torch.zeros(sizes[layer_i+1]) for layer_i in range(len(Us))] for step_i in range(steps)] for epoch_i in range(EPOCHS)]
    Us_track = [[torch.zeros((sizes[layer_i+1] * sizes[layer_i])) for layer_i in range(len(Us))] for epoch_i in range(EPOCHS)]
track_batch = 0

# Learning Loop
for ep in range(EPOCHS):
    # Track batches indexed: [layer][batch]
    batches_E_mags = [[] for _ in range(len(Us))]
    batches_R_mags = [[] for _ in range(len(Us))]
    batches_Us_mags = [[] for _ in range(len(Us))]
    batches_Us_stds = [[] for _ in range(len(Us))]

    # for idx in range(len(positive_dataset)):
    loop = tqdm(enumerate(dataloader), total=len(dataloader), leave=False)
    if ep > 0:
        temp_e = torch.zeros(len(Us))
        for i in range(len(Us)):
            temp_e[i] = epochs_E_mags[i][-1]
        loop.set_description(f"Epoch {ep+1}/{EPOCHS} - E: {temp_e.mean().item():.6f}")
    # loop = enumerate(dataloader)
    for batch_i, (x, y) in loop:
        x = x.flatten(start_dim=1).to(device)
        y = F.one_hot(y, num_classes=sizes[-1]).float().to(device)
        batch_size = x.shape[0]

        E = [torch.zeros((batch_size, sizes[i],)).to(device) for i in range(len(sizes)-1)]
        R = [torch.zeros((batch_size, sizes[i+1],)).to(device) for i in range(len(sizes)-1)]
        # R[-1] = y

        for step_i in range(steps):
            with torch.no_grad():
                E, R = step2(x, E, R, Us, Vs, y, corr_coeff)

            if step_i > 0:
                with torch.no_grad():
                    for i in range(len(Us)):
                        # print(f"shapes: {E[i].T.shape}, {R[i].shape}, {Us[i].weight.shape}")
                        Us[i].weight.grad = -(E[i].T @ R[i])/batch_size # -ve because we are performing grad descent, yet we want to increase connectivity 
                        if Us[i].bias is not None:
                            Us[i].bias.grad = Us[i].bias - x.mean(axis=0) if i == 0 else Us[i].bias - R[i-1].mean(axis=0) 
                        Vs[i].weight.grad = (R[i].T @ E[i])/batch_size # -ve because we are performing grad descent, yet we want to increase connectivity

                reg = reg_coeff * sum([U.weight.square().mean() for U in Us]) # L2
                # reg = reg_coeff * sum([U.weight.abs().mean() for U in Us]) # L1
                reg.backward()
                
                # Parameter Update
                for i in range(len(Us)):
                    Us[i].weight.data -= lr * Us[i].weight.grad
                    if Us[i].bias is not None:
                        Us[i].bias.data -= lr * Us[i].bias.grad
                    Vs[i].weight.data -= lr * Vs[i].weight.grad

                if batch_i == track_batch and track:
                    for i in range(len(Us)):
                        E_track[ep][step_i][i] = E[i].clone()
                        R_track[ep][step_i][i] = R[i].clone()
                        if step_i == 1:
                            Us_track[ep][i] = Us[i].weight.flatten().clone().detach()

        # # plot grads
        # if ep == 9:
        #     plt.figure(figsize=(15, 5))
        #     for i in range(len(Us)):
        #         plt.subplot(1, len(Us), i+1)
        #         plt.imshow(Us[i].weight.grad.cpu().numpy())
        #         plt.colorbar()
        #     plt.show()


        # Track batches
        temp_e = torch.zeros(len(Us))
        for i in range(len(Us)):
            batches_E_mags[i].append(E[i].norm(dim=1).mean().item())
            batches_R_mags[i].append(R[i].norm(dim=1).mean().item())
            # batches_Us_mags[i].append(Us[i].weight.data.abs().mean().item())
            batches_Us_mags[i].append(Us[i].weight.data.mean().item())
            batches_Us_stds[i].append(Us[i].weight.data.std().item())
        

    # Track epochs
    for i in range(len(Us)):
        epochs_E_mags[i].append(torch.tensor(batches_E_mags[i]).mean().item())
        epochs_R_mags[i].append(torch.tensor(batches_R_mags[i]).mean().item())
        epochs_Us_mags[i].append(torch.tensor(batches_Us_mags[i]).mean().item())
        epochs_Us_stds[i].append(torch.tensor(batches_Us_stds[i]).mean().item())

# # Plot weights
# plt.figure(figsize=(15, 5))
# for i in range(len(Us)):
#     plt.subplot(1, len(Us), i+1)
#     plt.imshow(Us[i].weight.data.cpu().numpy())
#     plt.colorbar()
# plt.show()

# Plot
fig, axs = plt.subplots(2,2, figsize=(8,8))
for i in range(len(Us)):
    axs[0,0].plot(epochs_E_mags[i], label=f"Layer {i}")
    axs[0,0].set_title('E_mags')
    axs[0,0].legend()
    axs[0,1].plot(epochs_R_mags[i], label=f"Layer {i}")
    axs[0,1].set_title('R_mags')
    axs[0,1].legend()
    axs[1,0].plot(epochs_Us_mags[i], label=f"Layer {i}")
    axs[1,0].set_title('Weight Mean')
    axs[1,0].legend()
    axs[1,1].plot(epochs_Us_stds[i], label=f"Layer {i}")
    axs[1,1].set_title('Weight Mean Std')
    axs[1,1].legend()
plt.show()
```

```python
def measure_errs(
        model,
        dataloader,
):
    E = [0.0 for _ in range(len(model.layers))]

    for batch_idx, (images, _) in enumerate(dataloader):
        images = images.flatten(start_dim=1)
        state = model(images, steps=5)
        for i, e in enumerate(state[0]):
            E[i] += e.abs().mean()
    
    for i in range(len(E)):
        E[i] /= batch_idx+1

    return E

def train(
        model,
        positive_dataset,
        negative_dataset,
        batch_size,
        lr,
        epochs,
        measure_errs_every=1,
        step=0,
):
    # writer = SummaryWriter("Deep_Learning/Experimental/Hebbian_Learning/mnist/out/logs")
    Es = [[] for _ in range(len(model.layers))]
    layer_reg_coeff = [0.3, 0.1]
    positive_dataloader = torch.utils.data.DataLoader(positive_dataset, batch_size=batch_size, shuffle=True)

    for epoch in range(epochs):
        negative_dataset = mix_dataset(negative_dataset)
        negative_dataloader = torch.utils.data.DataLoader(negative_dataset, batch_size=batch_size, shuffle=True)
        
        # For some dumb reason, i have to reinitialise the loop every epoch
        loop = tqdm(enumerate(positive_dataloader), total=len(positive_dataloader), leave=False) 

        neg_it = iter(negative_dataloader)
        for batch_idx, (images, y) in loop:
            images = images.flatten(start_dim=1)

            for phase in range(2):
                phase_coeff = 1.0
                if phase == 1: # Negative phase
                    continue
                    # images = torch.randn_like(images)
                    images = next(neg_it)[0].flatten(start_dim=1)
                    phase_coeff = -0.5

                state = model.init_state(len(images))
                for step_i in range(4):
                    state = model.step(images, state)
                    if step_i > 0:
                        model.zero_grad()
                        reg = -1.0 * sum(l.weight.square().mean() for l in model.layers)
                        reg.backward()
                        
                        for i in range(len(model.layers)):
                            model.layers[i].weight.grad += -state[0][i].T @ state[1][i]
                            model.layers[i].weight.grad /= len(images)
                            # print(f'epoch: {epoch}, batch: {batch_idx}, layer: {i}, phase: {phase}, step: {step_i}')
                            # print(f'e_mag: {state[0][i].abs().mean()}, r_mag: {state[1][i].abs().mean()}')
                            # print(model.layers[i].weight.grad.abs().mean())
                            model.layers[i].weight.data += phase_coeff*lr*model.layers[i].weight.grad

            step += len(images)

        if epoch % measure_errs_every == 0:
            E = measure_errs(model, positive_dataloader)
            for i, e in enumerate(E):
                Es[i].append(e)

    return Es, step
```
