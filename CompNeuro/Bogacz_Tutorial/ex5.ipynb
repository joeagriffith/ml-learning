```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
%matplotlib inline
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
dt = 0.01
max_t = 20
int(max_t / dt)
```

```python
phi_mean = 5
phi_var = 2
g_phi_ip1 = 5

trials = 1000
max_t = 2
dt = 0.1
lr = 0.01

Sigma = [1]

def simulate_network():
    phi = np.random.normal(phi_mean, phi_var)
    eps = sigmoid(np.random.normal(0, 1))
    e = sigmoid(np.random.normal(0, 1))
    for _ in range(int(max_t/dt)):
        eps += dt * (phi - g_phi_ip1 - e)
        e += dt * (Sigma[-1]*eps - e)
    return phi, eps, e

def train():
    for _ in range(trials):
        phi, eps, e = simulate_network()
        new_Sigma = Sigma[-1] + lr * (eps * e - 1)
        Sigma.append(new_Sigma)

train()

plt.plot(Sigma)

```
