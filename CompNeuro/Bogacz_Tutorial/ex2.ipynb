# Iterative Calculation of MAP

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
%matplotlib inline
```

```python
u = 2
Sigma_u = 1
v_p = 3
Sigma_p = 1

def g(v):
    return np.square(v)

def dg(v):
    return 2*v
```

```python
def simulate(max_t=5, dt=0.01, lr=0.01):
    ts = [0.0]
    phis = [v_p]

    while ts[-1] < max_t:
        eps = (v_p - phis[-1]) 
        pe = ((u - g(phis[-1])) * dg(phis[-1]))
        dFdPhi = eps / Sigma_p + pe / Sigma_u

        phis.append(phis[-1] + lr*dFdPhi)
        ts.append(ts[-1] + dt)
    
    return ts, phis

x, y = simulate()

plt.plot(x, y)
plt.show()
```
